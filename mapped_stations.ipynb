```python
import pandas as pd
import folium

# Read the cleaned data
df = pd.read_csv('/content/drive/MyDrive/Galveston_Hackathon/Cleaned_NOAA_Data.csv')

# Create a base map
m = folium.Map(location=[29.3, -94.8], zoom_start=10)

# Add data points to the map
for idx, row in df.iterrows():
    if not pd.isna(row['Latitude']) and not pd.isna(row['Longitude']):
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=5,
            color="blue",
            fill=True,
            fill_color="blue"
        ).add_to(m)

# Display the map
m
```
